```python
import numpy as np
import matplotlib.pyplot as plt

# 2. a)


def metEuler(v0, t0, tf, y0, dt):
    n = int((tf-t0) / dt + 0.1)

    t = np.zeros(n + 1)
    y = np.zeros(n + 1)
    vy = np.zeros(n + 1)
    ay = np.zeros(n + 1)

    vy[0] = v0
    t[0] = t0
    y[0] = y0

    for i in range(n):  
        ay[i] = -9.8 - ((-9.8/6.8*2) * vy[i] * np.abs(vy[i]))
        vy[i + 1] = vy[i] + ay[i] * dt
        y[i + 1] = y[i] + vy[i] * dt
        t[i + 1] = t[i] + dt

    return t, y, vy, ay

t,y,v,a = metEuler(50,0,5,0,5)

plt.plot(y,t)
plt.xlabel("Tempo (s)")
plt.ylabel("Altura (m)")
plt.show()
```
