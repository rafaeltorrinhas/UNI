```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
t0 = 0.0
tf = 2.0
dt = 0.001
v0 = 0.0
x0 = 0.0
M = 0.2
g = 9.80665


t = np.arange(t0,tf,dt)

a = np.zeros(np.size(t))

v = np.zeros(np.size(t))
v[0] = v0

x = np.zeros(np.size(t))
x[0] = x0

for i in range(np.size(t) - 1):

  a[i] = g if x[i] < 10 else 0.0 # usamos o operador de atribuição condicional

  # método de Euler-Cromer
  v[i+1] = v[i] + a[i] * dt
  x[i+1] = x[i] + v[i+1] * dt

plt.plot(t, x, "b-")
plt.xlabel("t [s]")
plt.ylabel("x [m]")
plt.show()
```

#### Velocidade final

```python
v1 = v
print(f"A veloocidade final é: {v1[-1]:.2f} m/s.")
```

#### Gráfico da Ep, Ec e Etotal em função do tempo

```python
y1 = np.where(x<10, 11.0-x, 1.0)
x1 = x
E_p1 = M * g * y1
E_c1 = 1/2 * M * v1 * v1

plt.plot(t,E_p1,"r-",t,E_c1,"g-",t,E_p1+E_c1,"b-")
plt.ylabel("t [s]")
plt.xlabel("Energia [J]")
plt.show()
```
