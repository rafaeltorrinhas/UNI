```python
# imports
import matplotlib.pyplot as plt
import numpy as np
```

```python
t0 = 0
tf = 10.0
dt = 0.001
v0 = 2 * np.pi
x0 = 1.0

G = 4.0 * np.pi ** 2
m = 3.003e-6

t = np.arange(t0,tf,dt)
a = np.zeros([np.size(t), 2])
v = np.zeros([np.size(t), 2])
r = np.zeros([np.size(t), 2])

v[0,:] = np.array([0, v0])
r[0,:] = np.array([x0, 0])

for i in range(np.size(t) - 1):
  a[i,:] = - G * r[i,:] / np.linalg.norm(r[i, :]) ** 3
  v[i+1, :] = v[i,:] + a[i,:] * dt
  r[i+1, :] = r[i,:] + v[i,:] * dt

plt.figure(figsize=(6,6))
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.plot(r[:,0], r[:,1], "b-")
plt.xlabel("r_x [AU]")
plt.ylabel("r_y [AU]")
plt.show()
```
