```python
t0 = 0.0
tf = 4.0
x0 = 0.0
v0 = 0.0
dt = 0.1

g = 9.8

t = np.arange( t0, tf, dt)
v = np.empty(np.size(t))
x = np.empty(np.size(t))

x[0] = x0
v[0] = v0

for i in range(np.size(t) - 1):
  v[i+1] = v[i] + g * dt
  x[i+1] = x[i] + v[i] * dt

i2 = int((2 - t0) / dt)
print(f"x(t = 2s) = {x[i2]} m")

```

# Exercício 2

```python
t0 = 0.0
tf = 4.0
y0 = 0.0
v0 = 10.0
vT = 100 * 1000 / 3600
dt = 0.0001
g = 9.8

D = g / vT**2

t = np.arange(t0, tf, dt)
y = np.empty(np.size(t))
v = np.empty(np.size(t))
a = np.empty(np.size(t))

a[0] = 0
v[0] = v0
y[0] = y0

for i in range(np.size(t) - 1):
    a[i] = - g - D * v[i] * np.abs(v[i])
    v[i+1] = v[i] + a[i] * dt
    y[i+1] = y[i] + v[i] * dt

plt.plot(t, y, "b-")
plt.plot(t, y0+v0*t-(0.5*g) * t**2, "r-")
plt.grid(axis="y", color="black", linewidth=0.5)
plt.xlabel("Tempo, t (s)")
plt.ylabel("Posição, y (m)")
plt.show()
```

```python
# indice e tempo para o qual a posição é máxima

imax = np.argmax(y)
tmax = t[imax]
print(f"Tempo correspondente à altura máxima, tmax = {tmax} m ")
print(f"Altura máxima, ymax = {y[imax]}, m")

# indice e tempo para o qual volta a passar por y=0
izero = np.size(y) - np.size(y[y<0])
tzero = t[izero]
print(f"Tempo de retorno à origem, tzero = {tzero} s")
```
