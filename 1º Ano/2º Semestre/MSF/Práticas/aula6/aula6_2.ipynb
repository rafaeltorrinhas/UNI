```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t0 = 0.0
tf = 1.0
dt = 0.01
t = np.arange(t0,tf,dt)
v_T = 100.0 * 1000 / 3600
g = 9.8
D = g / v_T ** 2
r0 = np.array([0.0,0.0,23.8])
v0 = np.array([25.0,5.0,-50.0])
w = 390.0
rho = 1.225
R = 0.11
A = np.pi * R ** 2
m = 0.45

a = np.zeros([3,np.size(t)])

v = np.zeros([3,np.size(t)])
v[:,0] = v0

r = np.zeros([3,np.size(t)])
r[:,0] = r0

for i in range(np.size(t) - 1):
  a[0, i] = -D * np.linalg.norm(v) * v[0, i] + A * rho * R * w * v[0,i] / (2 * m)
  a[1, i] = -g -D * np.linalg.norm(v) * v[1, i]
  a[2, i] = -D * np.linalg.norm(v) * v[0, i] - A * rho * R * w * v[0,i] / (2 * m)
  v[:, i + 1] = v[:, i] + a[:, i] * dt
  r[:, i + 1] = r[:, i] + v[:, i] * dt

plt.plot(r[0,:], r[1,:], 'b-')
plt.show()
```

```python
ixzero = np.size(r[0,:]) - np.size(r[0, r[0,:]<0])
txzero = t[ixzero]

print(f"txzero = {txzero}")
print(f"x = {r[0,ixzero]} m\ny ={r[1,ixzero]} m\nz = {r[2,ixzero]} m")
```
