```python
import numpy as np
import matplotlib.pyplot as plt
```

#### PÃªndulo Simples

```python
t0 = 0.0
tf = 10
dt = 0.001
g = 9.8

theta0 = np.radians(10) # 10 * np.pi / 180
w0 = 0.0

L = 1.0

t = np.arange(t0, tf, dt)
y = np.zeros(np.size(t))
w = np.zeros(np.size(t))

theta = np.zeros(np.size(t))

theta[0] = theta0
w[0] = w0

for i in range(np.size(t) - 1):
  y[i] = -g / L * np.sin(theta[i])
  w[i+1] = w[i] + y[i] * dt
  theta[i+1] = theta[i] + w[i + 1] * dt

plt.plot(t, np.degrees(theta), "-b")
plt.xlabel("Tempo decorrido, t [s]")
plt.ylabel("Ãngulo, theta [Âº]")
plt.show()
```

```python
T = 2*np.pi*np.sqrt(L / g)
print(f"PerÃ­odo de oscilaÃ§Ã£o, T = {T:.2f} s")
```
